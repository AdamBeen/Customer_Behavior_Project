```python
import pandas as pd

df = pd.read_csv("customer_shopping_behavior.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category') ['Review Rating'].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.lower()
df.head()
```

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df.columns
```

```python
#create a column to group customers by age 
labels = ['Young Adult','Adult','Middle-Aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column with the number of days between each purchase made by a customer
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Annually' : 365,
    'Every 3 months' : 91,
    'Bi-Weekly' : 14,
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
```
